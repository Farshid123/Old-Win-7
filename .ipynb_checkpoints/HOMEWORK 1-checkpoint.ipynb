```python
def pullText(index,what):
    leftInd = index - 50
    rightInd = index + 50
    theFound = what[leftInd:rightInd]
    return theFound
```

```python
print(pullText(10745,fullDialogue))
```

```python
def pullText2(index,what):
    leftInd = index - 50
    rightInd = index + 50
    theFound = what[leftInd:rightInd].replace("\r\n"," ")
    return theFound

print(pullText2(10745,fullDialogue))
```
